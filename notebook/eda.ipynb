5. Correlation & Statistical Analysis
- Which features are most correlated with churn?
- Are there multicollinearity issues?
- Are there outliers impacting churn?

```python
df["Churn_Status"].unique()
```

```python
df["Churn_Label"] = df["Churn_Status"].replace({"Exit": 1, "Retain":0})
```

```python
corr_matrix = df.drop(columns=["Customer_ID"]).select_dtypes(include="number").corr()
corr_matrix
```

```python
corr_matrix["Churn_Label"].sort_values(ascending=False)
```

```python
# Which features are most correlated with churn?
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
corr_matrix.columns
```

```python
# Are there multicollinearity issues?
threshold = 0.85
col_corr = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col_name = corr_matrix.columns[i]
            col_corr.add(col_name)
```

```python
col_corr
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)]
```

```python
# Are there outliers impacting churn?
balance_outliers = detect_outliers_iqr(df, "Account_Balance")
```

```python
balance_outliers
```

```python
plt.figure()
plt.boxplot(df["Account_Balance"])
plt.title("Account Balance Distribution with Outliers")
plt.ylabel("Balance")
plt.show()
```

```python
plt.figure()
plt.boxplot(df["Age"])
plt.title("Age Distribution with Outliers")
plt.ylabel("Age")
plt.show()
```

```python
num_cols = df.drop(columns=["Churn_Label"]).select_dtypes(include="number").columns

for col in num_cols:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col}: {len(outliers)} outliers")
```
