```python
import pandas as pd
import os
```

```python
bank_dataset = pd.read_csv("datasets/Bank_Churn.csv")
```

```python
bank_dataset.head()
```

```python
customer_info = pd.read_csv("datasets/CustomerInfo.csv")
```

```python
customer_info.head()
```

```python
merged_df = bank_dataset.merge(customer_info[["CustomerId", "Surname"]], 
                               on="CustomerId", how="left")
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
active_customer = pd.read_excel("datasets/ActiveCustomer.xlsx")
active_map = dict(zip(
    active_customer["ActiveID"],
    active_customer["ActiveCategory"]
))

credit_customer = pd.read_excel("datasets/CreditCard.xlsx")
credit_map = dict(zip(
    credit_customer["CreditID"],
    credit_customer["Category"]
))

exit_customer = pd.read_excel("datasets/ExitCustomer.xlsx")
exit_map = dict(zip(
    exit_customer["ExitID"],
    exit_customer["ExitCategory"]
))

gender_customer = pd.read_excel("datasets/Gender.xlsx")
gender_map = dict(zip(
    gender_customer["GenderID"],
    gender_customer["GenderCategory"]
))

geo_customer = pd.read_excel("datasets/Geography.xlsx")
geo_map = dict(zip(
    geo_customer["GeographyID"],
    geo_customer["GeographyLocation"]
))

```

```python
merged_df["IsActiveMember"] = merged_df["IsActiveMember"].map(active_map)
merged_df["HasCrCard"] = merged_df["HasCrCard"].map(credit_map)
merged_df["Exited"] = merged_df["Exited"].map(exit_map)
merged_df["GenderID"] = merged_df["GenderID"].map(gender_map)
merged_df["GeographyID"] = merged_df["GeographyID"].map(geo_map)
```

```python
merged_df.head()
```

```python

merged_df.info()
```

```python
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={
    "RowNumber": "Row_Number",
    "CustomerId": "Customer_ID",
    "CreditScore": "Credit_Score",
    "GeographyID": "Country",
    "GenderID": "Gender",
    "Age": "Age",
    "Tenure": "Tenure_Years",
    "Balance": "Account_Balance",
    "NumOfProducts": "Number_Of_Products",
    "HasCrCard": "Has_Credit_Card",
    "IsActiveMember": "Active_Status",
    "EstimatedSalary": "Estimated_Salary_EUR",
    "Exited": "Churn_Status",
    "Bank DOJ": "Bank_Join_Date",
    "Surname": "Customer_Surname"
})
```

```python
merged_df = merged_df[['Customer_ID', 
                       'Customer_Surname', 'Bank_Join_Date',
                       'Credit_Score', 'Country', 'Gender', 'Age',
       'Tenure_Years', 'Account_Balance', 'Number_Of_Products',
       'Has_Credit_Card', 'Active_Status', 'Estimated_Salary_EUR',
       'Churn_Status']].copy()
```
